2.  **Calculate Loss**: "I predicted X noise, but real noise was Y. How far off was I?" (MSE Loss).
3.  **Backward Pass**: Calculate gradients.
4.  **Optimizer Step**: Update the LoRA weights ($B \times A$) to reduce error.
