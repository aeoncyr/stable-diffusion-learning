Welcome to **Module 6: Building the App**. We use **Gradio** to build the UI.

### Why Gradio?
It is pure Python. No HTML/CSS/React needed.

### Core Concepts:
1.  **Interface**: High-level wrapper (Fn -> Inputs -> Outputs).
2.  **Blocks**: Low-level layout (Rows, Columns, Tabs).
3.  **Events**: Click, Change, Submit.

```python
# 1. Setup
import notebook_utils
project_root, device, dtype = notebook_utils.setup_notebook()

import gradio as gr
```

## 1. Hello World (Blocks API)

We will build a tiny app right inside this notebook.

```python
def reverse_text(text):
    return text[::-1]

with gr.Blocks() as demo:
    gr.Markdown("# My First AI App")
    
    with gr.Row():
        inp = gr.Textbox(placeholder="Type something...")
        out = gr.Textbox()
    
    btn = gr.Button("Reverse me!")
    
    # Verification Event
    btn.click(fn=reverse_text, inputs=inp, outputs=out)

print("App defined (Run the next cell to launch if outside a notebook, or just trust the syntax).")
```
